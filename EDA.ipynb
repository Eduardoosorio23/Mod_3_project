```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline
```

```python
df = pd.read_csv('Data/churm_data.csv')
df.head()
```

```python
df = df.drop(['phone number', 'state'], axis=1, inplace=False)
```

```python
df.isna().sum()#No Null in this data set
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.histplot(data=df, x="customer service calls", hue="churn", multiple="dodge", shrink=6)
```

```python
sns.histplot(data=df, x="total day charge", hue="churn", multiple="dodge")
```

```python
sns.histplot(data=df, x="international plan", hue="churn", multiple="dodge", shrink = .5)
```

```python
sns.histplot(data=df, x="total day minutes", hue="churn", multiple="dodge", shrink = 1)
```

```python
g = sns.JointGrid(data=df, x="total day charge", y="customer service calls", hue="churn")
# g.joint_plot(sns.scatterplot, sns.histplot)
g.plot_joint(sns.scatterplot, alpha=.5)
g.plot_marginals(sns.histplot, kde=True)
```

```python
h = sns.JointGrid(data=df, x="total day minutes", y="total day calls", hue="churn")
h.plot(sns.scatterplot, sns.histplot)
```

```python
j = sns.JointGrid(data=df, x="total night minutes", y="total night calls", hue="churn")
j.plot(sns.scatterplot, sns.histplot)
```

```python
df['international plan'] = pd.get_dummies(df['international plan'])
df['voice mail plan'] = pd.get_dummies(df['voice mail plan'])
df['churn'] = pd.get_dummies(df['churn'])
df.head(20)#in the churn column True = 0, False = 1
#in the international plan and voice mail plan columns, yes = 0, no = 1
```
